```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

# モーニングスターから国内上場企業の銘柄コードを取得してdfに変換してcsvにする関数
def extract_stock_num_from_monista():
    stock_num_dict = {'1000': ['1000', '1100', '1200', '1300', '1400', '1500', '1600', '1700', '1800', '1900'],
                     '2000': ['2000', '2100', '2200', '2300', '2400', '2500', '2600', '2700', '2800', '2900'],
                     '3000': ['3000', '3100', '3200', '3300', '3400', '3500', '3600', '3700', '3800', '3900'],
                     '4000': ['4000', '4100', '4200', '4300', '4400', '4500', '4600', '4700', '4800', '4900'],
                     '5000': ['5000', '5100', '5200', '5300', '5400', '5500', '5600', '5700', '5800', '5900'],
                     '6000': ['6000', '6100', '6200', '6300', '6400', '6500', '6600', '6700', '6800', '6900'],
                     '7000': ['7000', '7100', '7200', '7300', '7400', '7500', '7600', '7700', '7800', '7900'],
                     '8000': ['8000', '8100', '8200', '8300', '8400', '8500', '8600', '8700', '8800', '8900'],
                     '9000': ['9000', '9100', '9200', '9300', '9400', '9500', '9600', '9700', '9800', '9900'],}

    all_stock_num = []

    for num in stock_num_dict.keys():
        tmp_list = stock_num_dict[num]
        for subnum in tmp_list:
            target_url = 'http://portal.morningstarjp.com/StockInfo/sec/list?code={}&code_detail={}'.format(num, subnum)
            r = requests.get(target_url) # requests.getでwebから取得
            # 日本語が文字化けする→r.textではダメ　r.contentで渡してあげないと
            soup_perser = BeautifulSoup(r.content, 'html.parser')
            soup_lxml = BeautifulSoup(r.content, 'lxml') # 要素抽出するらしい
            # 全てのaタグを取得し、リンク先URL(href属性)を表示する
            for tmp in soup_lxml.find_all('a'):
                if tmp.get('href').startswith('/StockInfo/info/snap/') == True:
                    result = re.sub(r"\D", "", tmp.get('href'))
                    all_stock_num.append(result)
                    
    df = pd.DataFrame(all_stock_num)
    df.to_csv('all_stock_num.csv')
                    
    return all_stock_num
```

```python
import pandas as pd

# 株マップからファンダメンタル,クォンツスコアを取得する関数
def get_fundamental(all_stock_num_list: list):
    df = pd.DataFrame()
    for num in all_stock_num_list:
        # スクレイピング先URL
        factor_url = 'https://jp.kabumap.com/servlets/kabumap/Action?SRC=basic/factor/base&codetext={}'.format(num)
        compare_url = 'https://jp.kabumap.com/servlets/kabumap/Action?SRC=basic/compare/base&codetext={}'.format(num)
        # ページ取得」
        factor = pd.read_html(factor_url)
        compare = pd.read_html(compare_url)
        # 取得ページからファンダ、クオンツスコアのdfを作成、マージ
        fac = list(factor[9]['ファクター(項目名クリックでヘルプ)'])
        value = list(factor[9]['ファクター値'])
        name_tmp = compare[9].set_index(0).T.iloc[:, 0]
        stock_name = name_tmp[1]
        compare_df = compare[9].set_index(0).T.drop(compare[9].set_index(0).T.columns[0], axis=1)\
                        .rename(index={1: f'{num}({stock_name})'})
        factor_df = pd.DataFrame(value, index=fac).T.rename(index={0: f'{num}({stock_name})'})
        tmp_df = pd.concat([factor_df, compare_df], axis=1)

        if len(df) == 0:
            df = tmp_df
        else:
            df = pd.concat([df, tmp_df])
        print(num)
    return df
```

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import datetime
from urlextract import URLExtract

# 株マップから適時開示（決算短信）があるかないかを検知してpdfリンクを教えてくれる関数
def extract_kessan_stock(all_stock_num_list: list):
    extractor = URLExtract()
    stock_dict = dict()
    KESSAN = '決算'
    TODAY = datetime.datetime.now().strftime('%Y/%m/%d')

    for num in all_stock_num_list:
        urls = []
        news_url = 'https://jp.kabumap.com/servlets/kabumap/Action?SRC=basic/news/base&codetext={}'.format(num)
        r = requests.get(news_url)
        soup_lxml = BeautifulSoup(r.content, 'lxml')

        quote_elms = soup_lxml.find_all('div', class_='newsDate')
        if len(quote_elms) != 0:
            str_quote_elms = str(quote_elms[0])
            if (TODAY in str_quote_elms) and (KESSAN in str_quote_elms):
                urls.append(extractor.find_urls(str_quote_elms))
                stock_dict[num] = urls
            else:
                pass
#         else:
#             continue
            
#         if (TODAY in str_quote_elms) and (KESSAN in str_quote_elms):
#             urls.append(extractor.find_urls(str_quote_elms))
#             stock_dict[num] = urls
        else:
            pass
        print(num)
    return stock_dict
```

```python
import pandas as pd
# import seaborn as sns

# 株マップからテクニカルを抜き出す
def extract_technical(all_stock_num_list: list):
    stock_technical = dict()
    for num in all_stock_num_list:
        code = str(num)

        technical = pd.DataFrame()
        url_1 = 'https://jp.kabumap.com/servlets/kabumap/Action?SRC=techBunseki/base&codetext='
        url_2 = '&PERF_MODE='
        url = url_1 + code + url_2
        mode_num = ["0","1","2"]
        files = []
        for i in mode_num:
            technical = pd.read_html(url+i)
            if len(technical) == 11:
                continue
            technical_trend = technical[10]
            technical_overheated = technical[11]
            technical = technical[13]
            files.append(technical)
        if len(technical) != 11:
            files.append(technical_trend)
            files.append(technical_overheated)
            stock_technical[num] = files
        else:
            pass
#             stock_technical[num] = 'データなし'
        print(num)
    return stock_technical
```
